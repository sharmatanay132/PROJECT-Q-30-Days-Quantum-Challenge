```python
#Generation of random message bits
import random
def generate_bits(n):
    return [random.randint(0,1) for i in range(n)]
msg = generate_bits(8)
```

```python
#Generating random basis for Alice
def generate_bases(n):
    return [random.choice(['+','x']) for i in range(n)]
alice = generate_bases(8)
```

```python
#Generating random basis for Bob
bob = generate_bases(8)
```

```python
#Generation of Message
n = int(input('enter the no of bits of message :'))
print("message Bits :", msg)
print("Alice Bases:", alice)
print("Bob Bases :", bob)
```

```python
#Comparison of Basis between Alice and Bob
shared_key=[]
for i in range(n):
    if alice[i]==bob[i]:
        shared_key.append(msg[i])
print("Shared Secret Key")
print(shared_key)
```
